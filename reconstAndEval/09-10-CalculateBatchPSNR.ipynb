```python
import numpy as np
import glob
import pandas as pd
import tifffile
```

```python
def PSNR(gt, pred, range_= 65535.0 ):
    mse = np.mean((gt - pred)**2)
    return 20 * np.log10((range_)/np.sqrt(mse))

def zero_mean(x):
    return x-np.mean(x)

def fix_range(gt,x):
    a = np.sum(gt*x) / (np.sum(x*x) )
    return x*a

def fix(gt,x):
    gt_=zero_mean(gt)
    return fix_range(gt_, zero_mean(x)  )   

def bestPsnr(gt,pred):
    ra=(np.max(gt)-np.min(gt))/np.std(gt)
    #ra=(np.max(gt)-np.min(gt))/np.std(gt)
    gt_=zero_mean(gt)/np.std(gt)
    return PSNR(zero_mean(gt_), fix(gt_,pred),ra)
```

## Change these variables

```python
scheme = 'Noisy/' # Can be either of "Noisy/", "GT/", "N2V/", "N4V/", "PN2V/"
GTFiles=glob.glob('../19-08-2019/09-TifReconstructions/GT/*.tif')
GTFiles.sort()
print(len(GTFiles))
reconstructedFiles=glob.glob('../19-08-2019/09-TifReconstructions/'+scheme+'*.tif')
reconstructedFiles.sort()
print(len(reconstructedFiles))
export_file_path='../19-08-2019/10-Results/'+scheme
```

## Core

```python
gtImageList200=[]
bestPsnrList200=[]
reconstructionImageList200=[]
gtImageList400=[]
bestPsnrList400=[]
reconstructionImageList400=[]

for i in range(len(reconstructedFiles)-1):
    if(i%8 < 4):
        gtImage=tifffile.imread(GTFiles[i//8])
        reconstructionImage=tifffile.imread(reconstructedFiles[i])
        bestPsnrTemp=bestPsnr(gtImage, reconstructionImage)
        print(i, bestPsnrTemp)
        reconstructionImageList200.append(reconstructedFiles[i])
        bestPsnrList200.append(bestPsnrTemp)
        gtImageList200.append(GTFiles[i//8])
    else:
        gtImage=tifffile.imread(GTFiles[i//8])
        reconstructionImage=tifffile.imread(reconstructedFiles[i])
        bestPsnrTemp=bestPsnr(gtImage, reconstructionImage)
        print(i, bestPsnrTemp)
        reconstructionImageList400.append(reconstructedFiles[i])
        bestPsnrList400.append(bestPsnrTemp)
        gtImageList400.append(GTFiles[i//8])
```

## Average PSNR

```python
print("Average PSNR for signal 200", np.average(bestPsnrList200))
print("Average PSNR for signal 400", np.average(bestPsnrList400))
```
