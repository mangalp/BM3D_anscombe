```python
import numpy as np
from matplotlib import pyplot as plt
from tifffile import imread, imwrite
```

```python
def fastShuffle(series, num=1):
    length = series.shape[0]
    for i in range(num):
        series = series[np.random.permutation(length)]
    return series
```

```python
def shufflePixels(img):
    img_=img.copy()
    for i in range(9):
        print(i)
        for j in range(512):
            for k in range(512):
                np.random.shuffle(img_[:,i,j,k])
                #tmp=img_[:,i,j,k]
                #print(tmp.shape)
                #img_[:,i,j,k]=fastShuffle(tmp)
    return img_
```

```python
a= imread('20191107_568_488_oil1514_FluoBPAEcells_2DSIM_tl_C_lowpower-c2-sub.tif')
for i in range(16):
    sa= shufflePixels(a)
    imwrite('pixelShuffle/20191107_568_488_oil1514_FluoBPAEcells_2DSIM_tl_C_lowpower-c2-sub-shuffle'+str(i)+'.tif',sa)
```
